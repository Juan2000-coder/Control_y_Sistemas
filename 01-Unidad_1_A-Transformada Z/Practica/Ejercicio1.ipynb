## Z-transform and ROC

```python
x1_ = sm.Rational(1,3)**n
x2_ = sm.Rational(1,2)**n
x3_ = 2**n
```

### Señal 1

```python
x1 = x1_*sm.Heaviside(n, 1)
x1
```

```python
X1 = sm.Sum(x1 * z**(-n), (n, -sm.oo, sm.oo))
X1
```

```python
X1 = sm.Sum(x1_ * z**(-n), (n, 0, sm.oo))
X1.doit()
```

### Señal 2

```python
x2 = x2_*sm.Heaviside(-n-1, 1)
x2
```

```python
X2 = sm.Sum(x2 * z**(-n), (n, -sm.oo, sm.oo))
X2
```

```python
X2 = sm.Sum(-x2_ * z**(-n), (n, -sm.oo, -1))
X2.doit()
```
